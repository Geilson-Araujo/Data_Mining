```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import (accuracy_score, cohen_kappa_score,
                             confusion_matrix, classification_report)

# Seleção de atributos
from sklearn.feature_selection import VarianceThreshold

# Agregação de atributos
from sklearn.decomposition import PCA

# Biblioteca para tratamento de classe desbalanceada
from imblearn.over_sampling import RandomOverSampler, SMOTE

# Silencia mensagens de alerta do sklearn
import warnings
warnings.filterwarnings(action='ignore', module='sklearn')

# Configurações gerais
pd.set_option('display.max_columns', 200)
sns.set()
seed = 1234 # semente para pseudo-aleatoriedade e reprodução de resultados
```

## Carregamento da base de dados

```python
ibm = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition_Missing.csv')

# Exibe as cinco primeiras linhas do dataset
ibm.head()
```

```python
# Exibe um sumário da base de dados 
ibm.info(memory_usage='deep')
```

```python
# Tranforma todas as variáveis do tipo 'object' em 'category'
categorical_columns = ibm.select_dtypes(exclude=np.number).columns

for column in categorical_columns:
    ibm[column] = ibm[column].astype('category')
    
ibm.info(memory_usage='deep')  
```

## Tratamento de dados faltantes

---

Por meio do resumo acima, o atributo `BusinessTravel`, por exemplo, possui uma quantidade menor de linhas em relação às demais colunas que compõem a base de dados. Há, portanto, __dados faltantes__ em nossa base de dados.

Sendo assim, __identificaremos__ as colunas que possuem dados faltantes e, onde nos parecer viável, buscaremos alguma __forma de imputar__ esses dados. 

---

```python
# Exibe somente as colunas que possuem dados faltantes
# e a respectiva contagem de dados faltantes
missing_values = ibm.isnull().sum() # quantidade de dados faltantes por coluna
missing_values[missing_values != 0] 
```

```python
# Exibe o percentual de dados faltantes por atributo
missing_ratio = missing_values / len(ibm)
missing = pd.DataFrame(missing_ratio[missing_ratio != 0],
             columns=['missing_ratio'])
missing.style.format('{:.2%}')
```

```python
# Resumo Estatístico dos atributos numéricos que possuem dados faltantes
missing_columns = missing_ratio[missing_ratio != 0].index
ibm[missing_columns].describe().T
```

```python
# Exibe o tipo das variáveis dos atributos
ibm[missing_columns].dtypes
```

---   
1. A base de dados possui 5 (cinco) colunas com dados faltantes;     
2. O atributo `EmployeeCount` possui mais de 50% de seus dados faltantes e variância igual a 0 (zero). Logo, não irá acrescentar qualquer informação relevante ao nosso modelo. Portanto, podemos remover essa coluna da base de dados;     
3. Dentre os atributos que possuem dados faltantes, 4 (quatro) são do tipo numérico e 1 (um) categórico. Assim, utilizaremos técnicas específicas para cada caso.
---

```python
### Tratamento de missing values para o atributo "EmployeeCount"
# Remove a coluna 'EmployeeCount' da base de dados
ibm.drop('EmployeeCount', axis=1, inplace=True) 
missing_columns = missing_columns.drop('EmployeeCount') # Remove a feature 'EmployeeCount' dessa lista
```

### Variáveis numéricas

```python
# Exibe a média, mediana e quantidade
# dos atributos com dados faltantes,
# agrupados de acordo com a classe "Attrition"
ibm.groupby('Attrition')[missing_columns].agg(['mean', 'median', 'count'])
```

---
* O atributo `HourlyRate` possui média e mediana muito próximas, inclusive em relação às instâncias de classe. Parece-nos indiferente, portanto, utilizarmos um único valor - média, ou mediana -, substituindo __todos__ os dados faltantes dessa coluna;      
* Ademais, percebemos que os dados faltantes do tipo numérico, pertencem todos ao atributo de classe `No`, pois somente nele observamos redução na quantidade de dados.     

Abaixo, verificaremos a presença de _outliers_ nesses atributos a fim de optarmos pela utilização da média, ou mediana na substituição de dados faltantes.   

---

```python
# Exibe um boxplot, a fim de identificarmos graficamente a presença
# de outliers nos atributos numéricos com dados faltantes
ibm.loc[:, ['HourlyRate', 'JobLevel', 'MonthlyIncome']].boxplot();
```

---
Apesar de identificarmos a presença de _outliers_ em `MonthlyIncome`, as escalas das variáveis parecem bastante desproporcionais. Iremos normalizar esse subconjunto de dados pela média. Dessa forma, poderemos visualizar todos os [diagrams de caixa](https://pt.wikipedia.org/wiki/Diagrama_de_caixa) em uma mesma escala. 

---

```python
# Exibe boxplot dos atributos numéricos, normalizados pela média
df = ibm.loc[:, ['HourlyRate', 'JobLevel', 'MonthlyIncome']]
df_norm = df / df.mean() # normaliza o subconjunto de dados pela média
df_norm.boxplot();
```

---
Avaliando a quantidade de instâncias, medidas de posição, _outliers_ optamos por fazer a substituição desses valores faltantes pela __mediana, de acordo com a instâncias de classe__.

---

```python
# Tratamento de missing values das variáveis numéricas
# Substituição de valores faltantes pela mediana,
# de acordo com os atributos de classe
for column in ['HourlyRate', 'JobLevel', 'MonthlyIncome']:
    ibm[column] = ibm.groupby('Attrition')[column].transform(
        lambda series: series.fillna(series.median())
    )  
```

```python
# Exibe a média, mediana e quantidade das variáveis numéricas
# após tratamento de dados faltantes
ibm.groupby('Attrition')[df.columns].agg(['mean', 'median', 'count'])
```

### Variáveis categóricas

---
Identificaremos qual a maior frequência do atributo `BusinessTravel`, de acordo com a classe.

---

```python
# Exibe as frequências das variáveis do atributo por classe
ibm.groupby('Attrition')['BusinessTravel'].value_counts(dropna=False)
```

---
Nesse caso, a variável mais frequente é `Travel_Rarely`, independente do atributo de classe. Substituiremos, portanto, os valores faltantes dessa coluna por `Travel_Rarely`.

---

```python
# Substitui os valores faltantes pela moda
ibm.BusinessTravel.fillna(ibm.BusinessTravel.mode()[0], inplace=True)

# Exibe as frequências das variáveis de atributo 
# por classe, após o tratamento de dados faltantes
ibm.groupby('Attrition')['BusinessTravel'].value_counts(dropna=False)
```

```python
# Verifica se ainda há missing values no dataset
ibm.isnull().sum().any()
```

## Conversão de variáveis categóricas em numéricas

---

A fim de utilizarmos os modelos de Aprendizado de Máquina em Python, precisamos transformar todos os atributos categóricos de nossa base de dados em numéricos.

Há muitas formas de se realizar essa transformação. Utizaremos a técnica que transforma variáveis catégoricas em _dummy_.

---

```python
# Separa a base de dados entre atributos previsores e classe.
X = ibm.drop('Attrition', axis=1)
y = ibm.Attrition

# Converte as variáveis categóricas em numéricas, 
# utilizando a técnica de variáveis dummy
X = pd.get_dummies(X, drop_first=True)

# Exibe as cinco primeiras linhas do dataset
# após a criação de variáveis dummy
X.head()
```

## Seleção de atributos

### Análise de variância

```python
# Dimensões do dataset antes da seleção de atributos
X.shape
```

```python
# Instancia o modelo para seleção de atributos
selector = VarianceThreshold()                               
selector.fit(X)

# Realiza a seleção de atributos, removendo da base de dados
# todos aqueles com variância igual a 0 (zero)
mask = selector.get_support()
X = X.loc[:, mask]

# Dimensões do dataset após seleção de atributos
X.shape
```

### Análise de correlação
--- 
Atributos altamente correlacionados podem ser redundantes para o modelo. Podemos identificá-los por meio de uma __matriz de correlação__ e,  após análise, optar por manter na base de dados apenas um deles. 

---

```python
# Cria matriz de correlação entre os atributos da base de dados
matriz_de_correlacao = X.corr()


# Exibe graficamento a matriz de correlção, a fim de indetificarmos
# visualmente a presença, ou não, de atributos altamente correlacionados
fig, ax = plt.subplots(figsize=(25, 25))

sns.set(style='white') # define o background do gráfico

mask = np.triu(np.ones_like(matriz_de_correlacao, bool))

sns.heatmap(
    matriz_de_correlacao,
    cmap='RdBu',
    center=0,
    annot=True,
    fmt='.0%', # exibe os dados em valores percentuais
    linewidths=1,
    cbar=False,
    mask=mask,    
)

# Corrige um bug gerado pelo matplotlib
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)

plt.tight_layout()
plt.show()
```

---
Os atributos `MonthlyIncome` e `JobLevel` são altamente correlacionados, havendo uma lógica nesse resultado. Logo, optaremos por manter apenas um deles, reduzindo tanto redundâncias, quanto a complexidade do modelo. 

---

```python
### Tratamento de dados altamente correlacionados
# Exclui o atributo "JobLevel" da base de dados
X.drop('JobLevel', axis=1, inplace=True)

# Dimensões do dataset após seleção de atributos
X.shape
```

---

Passaremos a avaliar o comportamento de um modelo preditivo, a partir do tratamento de dados realizado.

Faremos uso do modelo [Support-vector machine](https://en.wikipedia.org/wiki/Support-vector_machine) (SVM), que tanto pode ser utilizado para classificação, quanto para regressão.

Entretanto, é imprescindível para esse tipo de modelo a __normalização de dados__.

---
## Normalização de dados

```python
# Separa a base de dados entre
# 70% para treino e 30% para teste
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    stratify=y,
    random_state=seed,
)

# Normalização dos dados
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train) # normalização da base de treino
X_test = scaler.transform(X_test) # aplica a normalização à base de teste
```

<div class="alert alert-block alert-danger">
<b>Atenção:</b> A normalização deve ser realizada com base nos dados de <b>treino</b>! E em seguida aplicada à base de teste. 
</div>

## SVM   
* Tratamento de dados faltantes;   
* Seleção de atributos por meio da variância e correlação;    
* Normalização.

```python
%%time

model = SVC(random_state=seed) # instancia o modelo
model.fit(X_train, y_train) # treina o modelo
y_pred = model.predict(X_test) # realiza previsão

# Matrix de confusão
matrix_de_confusao = pd.crosstab(y_test, y_pred, rownames=['Previsto'],
                                 colnames=['Real'], margins=True, )
# Resumo
print(
    f'{"Métricas":-^50s}',
    f'Acurácia do modelo: {accuracy_score(y_test, y_pred):.2%}',
    f'Kappa: {cohen_kappa_score(y_test, y_pred):.2%}\n',
    f'{"Matriz de Confusão":-^50}',
    matrix_de_confusao,
    f'\n{"Tempo de Processamento":-^50s}',
    sep='\n',
)
```

## Balanceamento de dados

```python
# Exibe a frequência de atributos de classe
y_train.value_counts()
```

---
Observamos que os atributos de classe estão significatimente desbalanceados.  

Assim sendo, utilizaremos duas técnicas de __balanceamento de dados__ a fim de melhoramos nosso modelo:   
* Over-sampling;  
* Synthetic Minority Over-Sampling Technique(SMOTE).  

---

### Over-sampling

```python
# Instancia o método de reamostragem 
# e gera novas bases de treino
# a partir daquelas já existentes
ros = RandomOverSampler(random_state=seed) 
X_res, y_res = ros.fit_resample(X_train, y_train)

# Exibe a frequência de atributos de classe
# após reamostragem
pd.Series(y_res, name='Attrition').value_counts()
```

<div class="alert alert-block alert-danger">
<b>Atenção:</b> O balanceamento deve ser realizado <b>somente</b> na base de dados de treino!
</div>

## SVM   
* Tratamento de dados faltantes;   
* Seleção de atributos por meio da variância e correlação;    
* Normalização;   
* Balanceamento de dados: Over-sampling.   

```python
%%time

model = SVC(random_state=seed) # instancia o modelo
model.fit(X_res, y_res) # treina o modelo com as bases de treino balaceadas
y_pred = model.predict(X_test) # realiza previsão com 
                               # a base de teste normalizada

# Matrix de confusão
matrix_de_confusao = pd.crosstab(y_test, y_pred, rownames=['Previsto'],
                                 colnames=['Real'], margins=True, )
# Resumo
print(
    f'{"Métricas":-^50s}',
    f'Acurácia do modelo: {accuracy_score(y_test, y_pred):.2%}',
    f'Kappa: {cohen_kappa_score(y_test, y_pred):.2%}\n',
    f'{"Matriz de Confusão":-^50}',
    matrix_de_confusao,
    f'\n{"Tempo de Processamento":-^50s}',
    sep='\n',
)
```

### SMOTE

```python
# Instancia o método de reamostragem 
# e gera novas bases de treino
# a partir daquelas já existentes
sm = SMOTE(random_state=seed, n_jobs=-1)

X_res, y_res = sm.fit_resample(X_train, y_train)

# Exibe a frequência de atributos de classe
# após reamostragem
pd.Series(y_res, name='Attrition').value_counts()
```

## SVM   
* Tratamento de dados faltantes;   
* Seleção de atributos por meio da variância e correlação;    
* Normalização;   
* Balanceamento de dados: SMOTE.   

```python
%%time

model = SVC(random_state=seed) # instancia o modelo
model.fit(X_res, y_res) # treina o modelo com as bases de treino balaceadas
y_pred = model.predict(X_test) # realiza previsão com 
                               # a base de teste normalizada

# Matrix de confusão
matrix_de_confusao = pd.crosstab(y_test, y_pred, rownames=['Previsto'],
                                 colnames=['Real'], margins=True, )
# Resumo
print(
    f'{"Métricas":-^50s}',
    f'Acurácia do modelo: {accuracy_score(y_test, y_pred):.2%}',
    f'Kappa: {cohen_kappa_score(y_test, y_pred):.2%}\n',
    f'{"Matriz de Confusão":-^50}',
    matrix_de_confusao,
    f'\n{"Tempo de Processamento":-^50s}',
    sep='\n',
)
```

---
A técnica de __over-sampling__ apresentou melhores resultados. Dessa forma, descartaremos a SMOTE.

---

```python
%%time

# Over-sampling
ros = RandomOverSampler(random_state=seed) 
X_res, y_res = ros.fit_resample(X_train, y_train)

model = SVC(random_state=seed) 
model.fit(X_res, y_res)
y_pred = model.predict(X_test)

matrix_de_confusao = pd.crosstab(y_test, y_pred, rownames=['Previsto'],
                                 colnames=['Real'], margins=True, )
print(
    f'{"Métricas":-^50s}',
    f'Acurácia do modelo: {accuracy_score(y_test, y_pred):.2%}',
    f'Kappa: {cohen_kappa_score(y_test, y_pred):.2%}\n',
    f'{"Matriz de Confusão":-^50}',
    matrix_de_confusao,
    f'\n{"Tempo de Processamento":-^50s}',
    sep='\n',
)
```

## Agregação de atributos
### Principal component analisys (PCA)

```python
X_res.shape
```

```python
from sklearn.decomposition import PCA
```

```python
%%time

pca = PCA(
    n_components=0.99, # Seleciona os atributos a fim de 
                       # manter 99% da variância acumulada
    random_state=seed
)

pca.fit(X_res)

X_pca = pca.transform(X_res) 
X_pca_test = pca.transform(X_test)

model = SVC(random_state=seed) # instancia o modelo
model.fit(X_pca, y_res) # treina o modelo com as bases de treino
                        # após pca e balaceadas
y_pred = model.predict(X_pca_test) # realiza previsão com 
                               # a base de teste normalizada

# Matrix de confusão
matrix_de_confusao = pd.crosstab(y_test, y_pred, rownames=['Previsto'],
                                 colnames=['Real'], margins=True, )
# Resumo
print(
    f'{"Métricas":-^50s}',
    f'Acurácia do modelo: {accuracy_score(y_test, y_pred):.2%}',
    f'Kappa: {cohen_kappa_score(y_test, y_pred):.2%}\n',
    f'{"Matriz de Confusão":-^50}',
    matrix_de_confusao,
    f'\n{"Dimensões da base de dados":-^50s}',
    f'Antes do PCA: {X_res.shape[0]} linhas e {X_res.shape[1]} colunas',
    f'Após o PCA: {X_pca.shape[0]} linhas e {X_pca.shape[1]} colunas',
    f'\n{"Tempo de Processamento":-^50s}',    
    sep='\n',
)
```
